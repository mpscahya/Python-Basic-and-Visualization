## A. Mathplotlib

https://matplotlib.org/

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Plot Basic

```python
x = np.linspace(0, 10, 10)
y = x**2
z = x + y

print(x)
print(y)
print(z)
```

```python
plt.plot(x,y)
plt.title('Try mafirst chart here')
plt.xlabel('Sumbu-X')
plt.ylabel('Sumbu-Y')
```

```python
plt.plot(x,x)
```

### Subplot

```python
plt.subplot(2, 3, 1) 
plt.plot(x,y)
plt.subplot(2, 3, 2)

plt.subplot(2, 3, 3)

plt.subplot(2, 3, 4)
plt.plot(x,x)

plt.subplot(2, 3, 5)

plt.subplot(2, 3, 6)

plt.tight_layout() # agar ada jarak tiap grafik
```

### Tipe Plot

```python
plt.subplot(2, 2, 1)
plt.plot(x,y)

plt.subplot(2 ,2 ,2)
plt.scatter(x,y)

plt.subplot(2, 2, 3)
plt.hist(x,y, color='green')

plt.subplot(2, 2, 4)
plt.fill(x,y)

plt.tight_layout()
```

### Legend

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, label='plot 1')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(y, x, label='plot 2')
plt.legend()
```

```python
# Gabung Chart
plt.plot(x, y , label= 'plot 1')
plt.plot(y, x, label= 'plot 2')
plt.legend()
```

### Object Oriented

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 0.5, 0.5])
 # x0,y0 = 0 ; x1,y1 = 0.5
axes.plot(x,y)
```

```python
fig = plt.figure()
axes_1 = fig.add_axes([0, 0, 0.5, 0.5])
axes_2 = fig.add_axes([0.6, 0, 0.5, 0.5])

axes_1.plot(x,y)
axes_2.scatter(x,y)
```

### Subplot(s) using Object Oriented

```python
fig, axes = plt.subplots(1, 2)

axes[0].plot(x,y)
axes[1].scatter(x,y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 4)

axes[0].plot(x, y)
axes[1].scatter(x, y)
axes[2].hist(x, y)
axes[3].fill(x, y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(x, y)
axes[0, 1].scatter(x, y)
axes[1, 0].hist(x, y)
axes[1, 1].fill(x, y)

plt.tight_layout()
```

### Figure Size ,Give Title, and Axis Name

**Figure Size in Object Oriented**

```python
fig = plt.figure(figsize=(8,6))
axes = fig.add_axes([0,0,0.5,0.5])
axes.plot(x,y)
axes.set_title('Plattern Curve Covid - 19 ')
axes.set_xlabel('Sumbu - X')
axes.set_ylabel('Sumbu - Y')
```

```python
fig = plt.figure(figsize=(8,6))
axes_1 = fig.add_axes([0, 0, 0.5, 0.5])
axes_2 = fig.add_axes([0.6, 0, 0.5, 0.5])

axes_1.plot(x, y)
axes_2.scatter(x, y)

axes_1.set_title('Plattern Curve Covid - 19')
axes_1.set_xlabel('Sumbu - X')
axes_1.set_ylabel('Sumbu - Y')

axes_2.set_title('Plattern Scatter Curve Covid - 19')
axes_2.set_xlabel('Sumbu - X')
axes_2.set_ylabel('Sumbu - Y')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8,7))

axes[0, 0].plot(x, y)
axes[0, 1].scatter(x, y)
axes[1, 0].hist(x, y)
axes[1, 1].fill(x, y)

axes[0, 0].set_title('Plattern Curve Covid - 19')
axes[0, 1].set_xlabel('Sumbu - X')
axes[1, 0].set_ylabel('Sumbu - Y')
axes[1, 1].set_xlabel('Sumbu - X')
axes[1, 1].set_ylabel('Sumbu - Y')

plt.tight_layout()
```

## C. Import Files

```python
data = pd.read_excel('/Users/user/Desktop/BootCamp/Superstore.xls')
```

```python
pd.set_option('display.max_columns',30)
```

```python
data.head()
```

```python
sns.distplot(data['Sales'])
```

### Trend Sales per waktu

```python
data_1 = data[['Order Date','Sales']].groupby('Order Date')['Sales'].sum().reset_index()
#reset_index = buat jadi data frame
```

```python
plt.figure(figsize=(20,5))
plt.plot(data_1.set_index('Order Date'))
plt.title('Kurva Penjualan Harian', fontsize=20)
plt.xlabel('Tanggal')
plt.ylabel('Sales')
```

## Trend Profit per waktu

```python
data_2 = data[['Order Date','Profit']].groupby('Order Date')['Profit'].sum().reset_index()
#reset_index = buat jadi data frame
```

```python
data_2
```
